# Análise das cores/raças declaradas pelos alunos de IES no Censo do Ensino Superior 2019

---

#### Começa importando as bibliotecas e pegando as informações desejadas do banco de dados.

```python
import pandas as pd
import random
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns


dataframe = pd.read_csv(
    'SUP_ALUNO_2019.csv',
    encoding = "ISO-8859-1", 
    usecols=['CO_IES', 'TP_COR_RACA'],
    sep='|'
    )

dataframe.head()
```

---
#### Aqui é inserido o código da Instituição de Ensino Superior e o seu nome para visualização nos gráficos:
##### Lista de códigos de algumas instituições de ensino:

```python
listaIES = {580 : 'UFPE', 
            409 : 'UPE', 
            587 : 'UFRPE', 
           1809 : 'IFPE',
           3984 : 'UNIVASF'}
```

#### Código da instituição que se deseja analisar:

```python
codigoIES = 409
nomeIES = listaIES[codigoIES]
```

#### Número total de alunos da instituição em análise:

```python
numeroEstudantesIES = dataframe.loc[dataframe['CO_IES'] == codigoIES].count()[0]
print(f"Quantidade de alunos da {nomeIES} avaliados pelo Censo: {numeroEstudantesIES} estudantes.")
```

---
### Gráfico para a visualização da porcentagem de raças declaradas pelos alunos de todas as IES brasileiras.

```python
numeroEstudantesBrasil = dataframe.shape[0]
estudantes = dataframe.loc[dataframe['CO_IES'] == codigoIES]

naoDeclarou = dataframe.loc[dataframe['TP_COR_RACA'] == 0].count()[0]
brancos = dataframe.loc[dataframe['TP_COR_RACA'] == 1].count()[0]
pretos = dataframe.loc[dataframe['TP_COR_RACA'] == 2].count()[0]
pardos = dataframe.loc[dataframe['TP_COR_RACA'] == 3].count()[0]
amarelos = dataframe.loc[dataframe['TP_COR_RACA'] == 4].count()[0]
indigenas = dataframe.loc[dataframe['TP_COR_RACA'] == 5].count()[0]
naoDispoeInformacao = dataframe.loc[dataframe['TP_COR_RACA'] == 9].count()[0]

naoInformado = naoDeclarou + naoDispoeInformacao

labels = ['Brancos(as)', 'Pretos(as)', 'Pardos(as)', 'Amarelos(as)', 'Indígenas', 'Não declarado']

sizes=[brancos*100/numeroEstudantesBrasil, 
       pretos*100/numeroEstudantesBrasil, 
       pardos*100/numeroEstudantesBrasil, 
       amarelos*100/numeroEstudantesBrasil, 
       indigenas*100/numeroEstudantesBrasil, 
       naoInformado*100/numeroEstudantesBrasil]

fig1, ax1 = plt.subplots(figsize=(24, 12))

plt.rcParams.update({'font.size': 14})

ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)

ax1.axis('equal')

ax1.set_title(f"Tipos de cor/raça declaradas pelos alunos do Brasil")

plt.show()
print(brancos, pretos, pardos, amarelos, indigenas, naoInformado)
print(f'Brancos: {brancos*100/numeroEstudantesBrasil:.1f}%'
      f'\nPretos: {pretos*100/numeroEstudantesBrasil:.1f}%'
      f'\nPardos: {pardos*100/numeroEstudantesBrasil:.1f}%' 
      f'\nAmarelos: {amarelos*100/numeroEstudantesBrasil:.1f}%'
      f'\nIndigenas: {indigenas*100/numeroEstudantesBrasil:.1f}%'
      f'\nNão informado: {naoInformado*100/numeroEstudantesBrasil:.1f}%')
```

---
### Gráfico para a visualização da porcentagem de raças declaradas pelos alunos de uma IES.

```python
estudantes = dataframe.loc[dataframe['CO_IES'] == codigoIES]

naoDeclarou = estudantes.loc[estudantes['TP_COR_RACA'] == 0].count()[0]
brancos = estudantes.loc[estudantes['TP_COR_RACA'] == 1].count()[0]
pretos = estudantes.loc[estudantes['TP_COR_RACA'] == 2].count()[0]
pardos = estudantes.loc[estudantes['TP_COR_RACA'] == 3].count()[0]
amarelos = estudantes.loc[estudantes['TP_COR_RACA'] == 4].count()[0]
indigenas = estudantes.loc[estudantes['TP_COR_RACA'] == 5].count()[0]
naoDispoeInformacao = estudantes.loc[estudantes['TP_COR_RACA'] == 9].count()[0]

naoInformado = naoDeclarou + naoDispoeInformacao

labels = ['Brancos(as)', 'Pretos(as)', 'Pardos(as)', 'Amarelos(as)', 'Indígenas', 'Não declarado']

sizes=[brancos*100/numeroEstudantesIES, 
       pretos*100/numeroEstudantesIES, 
       pardos*100/numeroEstudantesIES, 
       amarelos*100/numeroEstudantesIES, 
       indigenas*100/numeroEstudantesIES, 
       naoInformado*100/numeroEstudantesIES]

fig1, ax1 = plt.subplots(figsize=(24, 12))

plt.rcParams.update({'font.size': 14})

ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)

ax1.axis('equal')

ax1.set_title(f"Tipos de cor/raça declaradas pelos alunos do(a) {nomeIES}")

plt.show()

print(f'Brancos: {brancos*100/numeroEstudantesIES:.1f}%\nPretos: {pretos*100/numeroEstudantesIES:.1f}%\nPardos: {pardos*100/numeroEstudantesIES:.1f}%' 
      f'\nAmarelos: {amarelos*100/numeroEstudantesIES:.1f}%\nIndigenas: {indigenas*100/numeroEstudantesIES:.1f}%\nNão informado: {naoInformado*100/numeroEstudantesIES:.1f}%')
```
