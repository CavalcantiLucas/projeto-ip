# ***Análise das idades gerais dos alunos comparadas às dos alunos de medicina, na UFPE***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np

data = pd.read_csv(
    'SUP_ALUNO_2019.csv',
    encoding = "ISO-8859-1", 
    usecols=['NU_IDADE', 'CO_CURSO', 'CO_IES','TP_SEXO'],
    sep='|'
    )

data.head()
```

```python
students_ufpe = data.loc[data['CO_IES'] == 580]

course_med = students_ufpe.loc[students_ufpe['CO_CURSO'] == 13601]
```


# Plotagem dos gráficos com a concentração das idades por faixa etária

```python
sns.set_style('whitegrid')
ax = sns.kdeplot(students_ufpe.NU_IDADE,shade=True)
ax = sns.kdeplot(course_med.NU_IDADE,shade=True, color='r')
ax.set_ylabel('Concentração')
ax.set_xlabel('Idades')
plt.show()
```

## A curva em vermelho, que representa a concentração das idades dos alunos de medicina, é deslocada para a direita em relação à dos alunos de outros cursos, o que reflete a concorrência maior para a entrada no vestibular desse curso, como esperado.

```python
ages_med = course_med.loc[course_med['NU_IDADE']==23].count()[0]
print(ages_med)
```

## Perceba que a idade mais recorrente, para medicina, é 23 anos.

# ***Análise da quantidade de mulheres e homens nas Universidades***


## Essa análise foi feita com base no Censo Universitário de 2019, que ainda buscou dados considerando gêneros binários. Devido a isso, a análise englobará apenas as pessoas que se identificaram como homens ou mulheres.

```python
women = data.loc[data['TP_SEXO'] == 1].count()[0] #6928483
men = data.loc[data['TP_SEXO'] == 2].count()[0] #5422349

labels = ['Mulheres', 'Homens']
values = [women, men]

fig, ax = plt.subplots()

ax.pie(values, labels = labels, colors=['#FFE4E1', '#E0FFFF'], autopct='%.1f%%', shadow=False, startangle=90)

plt.title('Alunos de universidades no Brasil, em 2019, por sexo')

plt.show()
```

## Perceba que a quantidade de mulheres é maior, no geral. Embora não observemos uma diferença tão discrepante percentualmente, ela representa uma diferença de 1.506.134 alunas.

# Análise da quantidade de alunos homens e mulheres no Centro de Informática da UFPE

```python
cod_uni = 580
cod_cc = 13595
cod_ec = 58854
cod_si = 1136142

students = data.loc[data['CO_IES'] == cod_uni]

students_cc = students.loc[students['CO_CURSO'] == cod_cc]
students_ec = students.loc[students['CO_CURSO'] == cod_ec]
students_si = students.loc[students['CO_CURSO'] == cod_si]

women_cc = students_cc.loc[students_cc['TP_SEXO'] == 1].count()[0] #67
men_cc = students_cc.loc[students_cc['TP_SEXO'] == 2].count()[0] # 506

women_ec = students_ec.loc[students_ec['TP_SEXO'] == 1].count()[0] #93
men_ec = students_ec.loc[students_ec['TP_SEXO'] == 2].count()[0] #540

women_si = students_si.loc[students_si['TP_SEXO'] == 1].count()[0] #46
men_si = students_si.loc[students_si['TP_SEXO'] == 2].count()[0] #333

women_ufpe = students.loc[students['TP_SEXO'] == 1].count()[0] #18133
men_ufpe = students.loc[students['TP_SEXO'] == 2].count()[0] #18615
```

# Ciências da Computação

```python
plt.show()

labels = ['Mulheres', 'Homens']
values = [women_cc, men_cc]

fig, ax = plt.subplots()

ax.pie(values, labels = labels, colors=['#FFE4E1', '#E0FFFF'], autopct='%.1f%%', shadow=False, startangle=90)

plt.title('Ciências da Computação na UFPE, por sexo')

plt.show()
```

# Engenharia da Computação

```python
labels = ['Mulheres', 'Homens']
values = [women_ec, men_ec]

fig, ax = plt.subplots()

ax.pie(values, labels = labels, colors=['#FFE4E1', '#E0FFFF'], autopct='%.1f%%', shadow=False, startangle=90)

plt.title('Engenharia da Computação na UFPE, por sexo')

plt.show()
```

# Sitemas da Informação

```python
labels = ['Mulheres', 'Homens']
values = [women_si, men_si]

fig, ax = plt.subplots()

ax.pie(values, labels = labels, colors=['#FFE4E1', '#E0FFFF'], autopct='%.1f%%', shadow=False, startangle=90)

plt.title('Sistemas da Informação na UFPE, por sexo')

plt.show()
```

## Perceba que, unanimamente, a discrepância entre a concentração de homens e mulheres no Centro de Informática em 2019 era enorme em todos os cursos.

---
# Análise geral do centro


```python
women = women_cc + women_ec + women_si
men = men_cc + men_ec + men_si

labels = ['Mulheres', 'Homens']
values = [women, men]

fig, ax = plt.subplots()

ax.pie(values, labels = labels, colors=['#FFE4E1', '#E0FFFF'], autopct='%.1f%%', shadow=False, startangle=90)

plt.title('CIn, por sexo')

plt.show()
```

# Análise da UFPE

```python
women = women_ufpe
men = men_ufpe

plt.rcdefaults()
fig, ax = plt.subplots()

people = ('Mulheres', 'Homens')
y_pos = np.arange(len(people))
performance = [women, men]

ax.barh(y_pos, performance, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(people) 
ax.set_xlabel('Quantidade')
ax.set_title('Distribuição homens/mulheres')

plt.show()
```

## Percebemos que a UFPE tinha 482 mais homens que mulheres em 2019, de acordo com a predominância masculina no Centro de Informática, mas na contramão do país como um todo, que tem predominância feminina nas universidades.
